	1.Write a program that prints all numbers from 1 to 100, but:
	•Replace multiples of 3 with “Fizz”.
	•Replace multiples of 5 with “Buzz”.
	•Replace multiples of both with “FizzBuzz”.

```python
for n in range(1,100):
  if n == 3:
    print("Fizz")
  elif n == 5:
    print("Buzz")
  elif n%3 == 0 or n%5 == 0:
    print("FizzBuzz")
  else:
    print(n)
```

    2. Even or Odd 
    Ask the user for a number and print whether it’s even or odd.



```python
num = input('Please enter a number: ')

while True:
  try:
    if int(num)%2 == 0:
      print(f'{num} is an even number')
      break
    elif int(num)%2 == 1:
      print(f'{num} is an odd number')
      break
  except Exception as e:
    if not num.isdigit():
      print(f'Error! {e}')
      break
```



    3. Word Counter
    •   Ask for a sentence and count how many words it has.



```python
sentence =input('Enter a sentence: ')
words = sentence.split(" ")
count = 0
for i in words:
  count += 1
print(f'The number of words in this sentence is {count}')
```

    4. List Reverser 
        • Reverse a list manually (without .reverse() or slicing).

```python
list1 = [1, 2, 3, 4]
list2 = []

while list1:
    list2.append(list1.pop())
for i in list2:
  list1.append(i)
print(list1)
```

    5. Find Maximum 
        • Write a function that finds the largest number in a list.

```python
import random

def add_random_numbers():
  list1 = []
  for n in range(1,20):
    list1.append(random.randint(1,5000))
  print(list1)
  return list1

def find_numb():
  list1 = add_random_numbers()
  current_max = list1[0]
  for n in list1[1:]:
    if n > current_max:
      current_max = n
  print(f'The biggest value in the list is {current_max}')

find_numb()
```

    6. Track the Minimum Number (loop version)

    Write a function that: 
    • generates 10 random numbers between 1 and 100 
    • loops through them to find the smallest one (without using min()) 
    • prints both the list and the smallest number


```python
import random

def add_rand():
  list1 = []
  for n in range(1, 10):
    list1.append(random.randint(1, 200))
  print (list1)
  return list1

def find_min():
  list1 = add_rand()
  current_min = list1[0]
  for n in list1[1:]:
    if n < current_min:
      current_min = n
  print(f'The smallest value in the list is {current_min}')

find_min()
```

    7. Count Greater Than
    Write a function that: 
    • creates a list of 10 random numbers between 1 and 50 
    • counts how many numbers are greater than 25 
    • prints both the list and the count


```python
import random

def add_numb():
  list1 = []
  for i in range(1,10):
    list1.append(random.randint(1,50))
  print(list1)
  return list1

def greater():
  list1 = add_numb()
  count = 0
  for n in list1:
    if n > 25:
      count += 1
  print(f'The list contains {count} values that are greater than 25')

greater()
```
